```python
import pandas as pd
import numpy as np
from pymongo import MongoClient
from transformers import pipeline
```

```python
client = MongoClient('mongodb://localhost:27017')
db = client['StockData']
collection = db['News']
```

```python
cursor = collection.find({'Stock_symbol': "SPY"}, {'_id': 0}).allow_disk_use(True).batch_size(100)  # Exclude _id field
data = list(cursor)
df = pd.DataFrame(data)
```

```python
stock_data = pd.read_csv('C:/Users/kevin/Downloads/ML/full_history/full_history/SPY.csv')
df = df.rename(columns={'Date': 'date'})
df.head(5)
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.date
stock_data['date'] = pd.to_datetime(stock_data['date'])
stock_data['date'] = stock_data['date'].dt.date
merged_df = pd.merge(stock_data, df, on='date', how='inner')
merged_df.head()
```

```python
#pipe = pipeline("text-classification", model="lxyuan/distilbert-base-multilingual-cased-sentiments-student")
finpipe= pipeline("text-classification", model="mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis")
```

```python
print(merged_df.Lsa_summary[7])
finpipe(merged_df.Lsa_summary[7])
```

```python
def get_sentiment_score(summary):
    output = (finpipe(summary)[0])
    sentiment = output.get('label')
    score = output.get('score')
    if sentiment == 'positive' :
        return score
    else :
        return -score
```

```python
# Convert column to string type
merged_df['Lsa_summary'] = merged_df['Lsa_summary'].astype(str)

# Remove any missing values
merged_df = merged_df.dropna(subset=['Lsa_summary'])

merged_df['sentiment_score'] = merged_df['Lsa_summary'].apply(lambda x: get_sentiment_score(x))
```

```python
merged_df.head()
```

```python

# Calculate the difference between Open and Close prices
merged_df['Price_Diff'] = merged_df['open'] - merged_df['close']

# Calculate the label based on Price_Diff
merged_df['Label'] = merged_df['Price_Diff'].apply(lambda x: 1 if x > 0 else 0)

# Drop the Price_Diff column if not needed
merged_df.drop('Price_Diff', axis=1, inplace=True)


merged_df.head()
```

```python
keep_columns = ['open', 'high', 'low', 'close', 'volume', 'sentiment_score', 'Label']
df = merged_df[keep_columns]
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
x= df
x = np.array(df.drop(['Label'], axis=1)) 
y= np.array(df['Label'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state =0)
```

```python
model = LinearDiscriminantAnalysis().fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
predictions
```

```python
print( classification_report(y_test, predictions))
```
